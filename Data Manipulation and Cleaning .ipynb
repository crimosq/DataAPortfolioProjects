```python
# importing modules (pandas, numpy, matplotlib.pyplot, seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing libraries

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
#url datset and passing it to uropen() to get html of page

url="http://www.hubertiming.com/results/2017GPTR10K"
html =urlopen(url)
```

```python
#Passing html to the BeautifulSoup() function
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
#Get the title

title = soup.title
print(title)
```

```python
#Print out the text
text = soup.get_text()
```

```python
soup.find_all('a')
```

```python
#for loop to get ('href') method to extract & only print out hyperlinks
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
    
```

```python
#print out table rows only, pass the 'tr' argument in soup.find_all()

rows = soup.find_all("tr")
print(rows[:10])
```

```python
# for loop that iterates through table rows and prints out the cells of rows

for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)

```

```python
#BeutifulSoup() & get_text() method to extract the text without html tags

str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
#generates an empty listt, extracts text between html tags for each row and append it to assigned list

import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean,'', str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
#converting list into a dataframe using Pandas
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
# DATA MANIPULATIONS AND CLEANING
```

```python
#splitting the "0" column using str.plit() method
```

```python
df1 = df[0].str.split(',', expand= True)
#df1.head(10)
```

```python
# strip() method to remove the opening square bracket on column "0."

df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
#using find_all() method to get the table headers
col_labels= soup.find_all("th")
```

```python
#Beautiful soup to extract text between html tags for table headers
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str,'lxml').get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
#converting list of heraders into a pandas dataframe
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
#split column "0" into multiple columns at the comma position for all rows
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
#concatenate 2 data frames using concat() method
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
#assigning the first row to be the table header
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
#Overview of the data
df5.info()
df5.shape
```

```python
#droping rows with any missing values
df6 = df5.dropna(axis=0, how='any')
```

```python
#table header is replicated as the first row in df5, dropping it as well
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
#more data cleaning 
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.rename(columns={' Team]': 'Team'},inplace=True)
df7.head()
```

```python
df7['Team'] = df7['Team'].str.strip(']')
df7.head()
```
